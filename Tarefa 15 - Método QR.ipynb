**Algoritimo do MÃ©todo de QR**

```python
import numpy as np
```

```python
def qr(A, erro = 1e-8, imprime=False ,k_max = 100000):
    n = A.shape[0]
    A_velha = np.copy(A)
    A_nova = np.copy(A)
    P = np.eye(n)
    val = 100
    k = 1
    while val > erro and k < k_max:
        Q, R = decomposicaoQR(A_velha)
        if imprime:
            print('-'*20)
            print('A{}={}'.format(k, (R@Q).round(3)) )
            print('-'*20)
            
        A_nova = R@Q
        
        A_velha = np.copy(A_nova)
        P = P@Q
        
        val = soma_diagonal(A_nova)
        k += 1
    
    lamb = np.diag(A_nova).reshape(n, 1)
    return P, lamb
```

```python
def soma_diagonal(A):
    n = A.shape[0]
    s = 0
    for i in range(n):
        for j in range(n):
            if i > j:
                s += A[i, j]**2
    return s
```

```python
def decomposicaoQR(A):
    n = A.shape[0]
    QT = np.eye(n)
    R_velha = np.copy(A)
    R_nova = np.copy(A)
    
    for j in range(0, n-1):
        for i in range(j+1, n):
            Jij = matriz_jacobi(R_velha, i , j)
            
            R_nova = Jij @ R_velha
            
            R_velha = np.copy(R_nova)
            
            QT = Jij@QT
    
    return QT.T, R_nova
```

```python
def matriz_jacobi(A, i , j):
    theta = 0
    erro = 1e-6
    n = A.shape[0]
    Jij = np.eye(n)
    if np.abs(A[i, j]) <= erro:
        return Jij
    if np.abs(A[j, j]) <= erro:
        if np.abs(A[i, j]) < 0:
            theta = np.pi / 2
        else:
            theta = - np.pi / 2
    else:
        theta = np.arctan( -A[i, j] / A[j, j] )
        
    s = np.sin(theta)
    c = np.cos(theta)
    Jij[i, i] = c
    Jij[j, j] = c
    Jij[i, j] = s
    Jij[j, i] = -s
    
    return Jij
```
